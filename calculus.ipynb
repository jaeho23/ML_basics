```python
import sympy as sym
x=sym.Symbol('x')
f=x**5
y=sym.diff(f)
print(y)
print(y.subs(x,2))
```

```python
fc=sym.cos(x)
yc=sym.diff(fc)
print(yc)
```

```python
def numerical_derivative(f,x):
  delta_x=1e-4
  return (f(x+delta_x)-f(x-delta_x))/(2*delta_x)

def my_func1(x):
  return x**2

result=numerical_derivative(my_func1,3)
print(result)
```

```python
x=sym.Symbol('x')
f=(x**2+1)*sym.cos(x)
y=sym.diff(f)
print(y)
```

```python
x=sym.Symbol('x')
p=sym.exp(x)-1
h=-1/3*p**2+p+1/5
y=sym.diff(h)
print(y)
```

```python
x,y,z=sym.symbols('x y z')
f=sym.exp(y*z**2)*sym.sin(x)
fx=sym.diff(f,x)
fy=sym.diff(f,y)
fz=sym.diff(f,z)
print(fx, fy, fz)
```

```python
t=sym.Symbol('t')
x=t-1
y=t**2
z=1/t
f=sym.exp(y*z**2)*sym.sin(x)
ft=sym.diff(f)
ft
```

```python
t=sym.Symbol('t')
f=t**5-5*t**3+4*t
d1=sym.Derivative(f,t).doit()
print(d1)
d2=sym.Derivative(f,t,2).doit()
print(d2)
```

```python
critical_points=sym.solve(d1)
A=critical_points[0];B=critical_points[1];C=critical_points[2];D=critical_points[3];
print(A.evalf(),B.evalf(),C.evalf(),D.evalf())
```
