```python
import pandas as pd

df=pd.read_csv("drive/MyDrive/Colab Notebooks/drug200.csv")
df
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import train_test_split

X = df.iloc[:, :5]
Y = df.iloc[:, 5]

# 데이터 전처리
X.loc[df.Sex=='F','Sex']=1
X.loc[df.Sex=='M','Sex']=0
X.loc[df.BP=='HIGH','BP']=2
X.loc[df.BP=='NORMAL','BP']=1
X.loc[df.BP=='LOW','BP']=0
X.loc[df.Cholesterol=='HIGH','Cholesterol']=2
X.loc[df.Cholesterol=='NORMAL','Cholesterol']=1

X_scaled= MinMaxScaler().fit_transform(X)

# 훈련셋, 시험셋 나누기
train_input, test_input, train_target, test_target = train_test_split(X_scaled, Y, test_size=0.1, random_state=35)
train_input, val_input, train_target, val_target = train_test_split(train_input, train_target, test_size=0.2, random_state=35)

# 모델 구축
sc = SGDClassifier(loss='log',max_iter=10, random_state=35)
sc.fit(train_input, train_target)
print(sc.score(train_input, train_target))
print(sc.score(test_input, test_target))
```
