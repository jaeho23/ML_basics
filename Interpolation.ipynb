```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.interpolate as interpolate

line=[]
n=50

for i in range(0,n):
  sample=np.linspace(-1.5*np.pi,1.5*np.pi,n)
  if sample[i]<=2 or sample[i]>=3:
    line.append(sample[i])
  elif i%5==0:
    line.append(sample[i])

x=line
y=np.sin(x)

plt.plot(x,y,'k*')

f=interpolate.interp1d(x,y,kind='nearest')
xi=np.linspace(-1.5*np.pi,1.5*np.pi,200)
yi=f(xi)
plt.plot(xi,yi,'r--')

fc=interpolate.interp1d(x,y,kind='cubic')
xc=np.linspace(-1.5*np.pi,1.5*np.pi,200)
yc=fc(xc)
plt.plot(xc,yc,'b--')
```

```python
x=[1,2,3,4,5]
y=[26.5,21.8,14.5,6.5,-1.3]
plt.plot(x,y,'k*')

fl=interpolate.interp1d(x,y,kind='linear')
xli=np.linspace(1,5,200)
yli=fl(xli)
plt.plot(xli,yli,'r--')

x_r=[1.3,3.7]
y_r=[fl(1.3),fl(3.7)]
plt.scatter(x_r,y_r,c='b',s=100)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(labels=('data','linear','projected'))
print(y_r)
```

```python
x=[1,2,3,4,5]
y=[26.5,21.8,14.5,6.5,-1.3]
plt.plot(x,y,'k*')

fl=interpolate.interp1d(x,y,kind='linear')
xli=np.linspace(1,5,200)
yli=fl(xli)
#plt.plot(xli,yli,'r--')

fla=interpolate.lagrange(x,y)
xla=np.linspace(1,5,200)
yla=fla(xla)
plt.plot(xla,yla,'r--')

x_r=[1.3,3.7]
y_r=[fla(1.3),fla(3.7)]
plt.scatter(x_r,y_r,c='b',s=100)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(labels=('data','lagrange','projected'))
print(y_r)
```

```python
n=31
nn=int((n-1)/2)
x=np.linspace(-1,1,n)
y=x/abs(x)
y[nn]=0
plt.plot(x,y,'k*')

fl=interpolate.interp1d(x,y,kind='linear')
xl=np.linspace(-1,1,n)
yl=fl(xl)
plt.plot(xl,yl,'r')

fla=interpolate.lagrange(x,y)
xla=np.linspace(-1,1,n*2)
yla=fla(xla)
plt.plot(xla,yla,'b')
```

```python
x=[1,2,3,4,5]
y=[26.5,21.8,14.5,6.5,-1.3]
f1=interpolate.interp1d(x,y,kind='nearest')
f2=interpolate.interp1d(x,y,kind='zero')
f3=interpolate.interp1d(x,y,kind='next')
f4=interpolate.lagrange(x,y)

x1=np.linspace(1,5,100)
y1=f1(x1);y2=f2(x1);y3=f3(x1);y4=f4(x1)

plt.plot(x,y,'k*',x1,y1,'--',x1,y2,x1,y3,x1,y4)
plt.xlabel('x');plt.ylabel('y')
plt.legend(labels=('data','nearest','zero','next','lagrange'))
```
