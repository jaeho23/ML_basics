```python
import pandas as pd
import numpy as np
from tabulate import tabulate

junc_data_56=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/1. DataCon#1_Target_Junction_Pressure(56days).csv", header=None)
print(tabulate(junc_data_56))
```

```python
import datetime as dt

x = dt.datetime(2022, 1, 1)
dx = dt.timedelta(days=0, hours=0, minutes=5)

junctions = ['J2', 'J6', 'J7', 'J9', 'J11']
junc_data_56.columns=junctions

print(junc_data_56)
```

```python
y=[]

for i in range(0,16129):
  y.append(x+dx*i)

del y[0]

print(y)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.plot(y, junc_data_56)
plt.xlabel('time')
plt.ylabel('junction data')

plt.show()
```
